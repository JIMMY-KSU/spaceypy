```python
## Download ephemeris if you haven't done so already
from poliastro import ephem
from astropy import time
ephem.download_kernel("de421")

## Make sure this works (might require restart if running for the first time after downloading)
epoch = time.Time("2015-05-09 10:43")
r, v = ephem.planet_ephem(ephem.EARTH, epoch)
```

```python
## Let's make a porkchop plot
depart_start = "2016-01-01"
depart_end   = "2016-04-01"
arrive_start = "2016-06-01"
arrive_end   = "2016-12-01"

# Returns departure and arrival position and velocity vectors
rd,ra,vd,va = astro.porkchop("Earth","Mars",depart_start, depart_end, arrive_start, arrive_end)

#porkchop() wraps a lot of sub utilities. See actual code for example usage
```
